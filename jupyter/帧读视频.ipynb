```python

# -*- encoding: utf-8 -*-

import cv2
import os
import time
import random

images = './fish/'

if not os.path.exists(images):

    os.mkdir(images)

 

cap = cv2.VideoCapture("2020-12-14_10_47_15.mp4")

c=0

while(1):

    success, frame = cap.read()

    if success:
        number = random.random()
        if(number>0.0008):
            continue
        else:
            print(number)
            time.sleep(number)
            cv2.imwrite(images+str(c) + '.jpg',frame)
            c=c+1
    else:

        break

cap.release()
```
